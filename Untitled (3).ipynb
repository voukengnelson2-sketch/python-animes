```python
import pandas as pd

# 1. Chargement du fichier
df = pd.read_csv("animes (2).csv")

# 2. Nettoyage
# Supprime les doublons
df_clean = df.drop_duplicates()
# Supprime les lignes avec des cases vides
df_clean = df_clean.dropna()

# 3. Affichage TOTAL (sans cacher de lignes)
pd.set_option('display.max_rows', None)

# Afficher le tableau complet
display(df_clean)
```

```python
# Filtrer les anim√©s avec une note sup√©rieure ou egal 9
excellents = df[df['Note_Globale'] >= 9]

print(f"üåü {len(excellents)} anim√©s ont une note >= 9 :")
print(excellents[['Anime', 'Note_Globale']].to_string(index=False))
```

```python
# Anim√©s avec note > 9 ET qui sont termin√©s
condition1 = df['Note_Globale'] > 9
condition2 = df['Status'] == 'Fini'

bons_finis = df[condition1 & condition2]

print(f"‚úÖ {len(bons_finis)} anim√©s avec note > 9 ET termin√©s :")
print(bons_finis[['Anime', 'Note_Globale', 'Status']].head(10).to_string(index=False))
```

```python
# Moyenne des notes
note_moyenne = df['Note_Globale'].mean()

print(f"üìä Note moyenne des anim√©s : {note_moyenne:.2f} / 10")

# Moyenne du nombre d'√©pisodes
episodes_moyen = df['Nb_Episodes'].mean()
print(f"üì∫ Nombre moyen d'√©pisodes : {episodes_moyen:.0f}")
```

```python
# Note la plus haute et la plus basse
note_max = df['Note_Globale'].max()
note_min = df['Note_Globale'].min()

print(f"‚≠ê Note la plus haute : {note_max}")
print(f"üìâ Note la plus basse : {note_min}")

# Quel anim√© a la meilleure note ?
meilleur = df[df['Note_Globale'] == note_max]['Anime'].values[0]
print(f"\nüèÜ Le mieux not√© : {meilleur}")
```

```python
import matplotlib.pyplot as plt

# 1. On compte les statuts sur tes donn√©es PROPRES
# (V√©rifie bien si ta colonne s'appelle 'Status' ou 'Statut')
counts = df_clean['Status'].value_counts()  # Mets 'Statut' si erreur

# 2. Cr√©ation du graphique
plt.figure(figsize=(8, 5))

# Liste de couleurs sympas
couleurs = ['#22c55e', '#6366f1', '#f59e0b', '#ef4444']

# On dessine les barres
# L'astuce ici : on s'assure qu'il y a assez de couleurs, sinon on r√©p√®te la liste
bars = plt.bar(counts.index, counts.values, color=couleurs * (len(counts) // len(couleurs) + 1))

# 3. Ajouter les chiffres au-dessus des barres (tr√®s utile !)
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.5,
             f'{int(height)}', ha='center', fontweight='bold', fontsize=12)

# Titres
plt.title("üìä R√©partition de mes anim√©s (Fini vs En cours)", fontsize=14, fontweight='bold')
plt.ylabel("Nombre d'anim√©s")

# Enlever la bordure du haut et de droite pour faire plus "pro"
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.show()
```

```python
# Nuage de points
plt.figure(figsize=(10, 6))

# Limiter √† 500 √©pisodes max pour mieux voir
df_filtre = df[df['Nb_Episodes'] <= 500]

plt.scatter(df_filtre['Nb_Episodes'], df_filtre['Note_Globale'], 
           alpha=0.6, c='#6366f1', s=80, edgecolor='white')

plt.title("üìà Note vs Nombre d'√©pisodes", fontsize=14, fontweight='bold')
plt.xlabel("Nombre d'√©pisodes")
plt.ylabel("Note globale")
plt.grid(True, alpha=0.3)
plt.show()

print("üí° Chaque point = un anim√©")
```

```python
# Corr√©lation Note Globale vs Note du Pire √âpisode
corr_pire = df['Note_Globale'].corr(df['Note_Pire_Ep'])

print(f"üîó Corr√©lation Note Globale vs Note Pire Ep : {corr_pire:.2f}")

# Comparaison
print(f"\nüìä R√©capitulatif :")
print(f"   ‚Ä¢ Note Meilleur Ep : forte corr√©lation positive")
print(f"   ‚Ä¢ Note Pire Ep    : corr√©lation plus faible")
print(f"\nüí° Un anim√© peut avoir un mauvais √©pisode et rester bien not√© !")
```

```python
# 1. Calcul des corr√©lations sur tes donn√©es propres
corr_meilleur = df_clean['Note_Globale'].corr(df_clean['Note_Meilleur_Ep'])

# (Est-ce que la note globale est plomb√©e par le pire √©pisode ?)
corr_pire = df_clean['Note_Globale'].corr(df_clean['Note_Pire_Ep'])

# 2. Affichage des r√©sultats
print(f" Corr√©lation avec le MEILLEUR √©pisode : {corr_meilleur:.2f}")
print(f" Corr√©lation avec le PIRE √©pisode     : {corr_pire:.2f}")

# 3. Interpr√©tation automatique
print(f"\n Ce que √ßa veut dire pour tes donn√©es :")
if corr_meilleur > corr_pire:
    print("  La note globale d'un anim√© d√©pend surtout de ses MOMENTS FORTS.")
    print("  (Les gens pardonnent les mauvais √©pisodes si les bons sont incroyables.)")
else:
    print("  La note globale est tr√®s sensible aux MAUVAIS √©pisodes.")
    print("  (Un rat√© fait vite baisser la moyenne.)")

print("\n Rappel : Plus le chiffre est proche de 1, plus le lien est fort !")
```
